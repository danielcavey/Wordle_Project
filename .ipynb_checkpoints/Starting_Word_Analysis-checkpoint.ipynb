```python
# Modules that need to be imported

import numpy as np
import string
from itertools import product
```

```python
# function that determines the colours that a letter causes in each position

def letter_scoring(ans,letter):
    score = np.array(["black"]*5)
    if letter in ans:
        score = np.array(["yellow"]*5)
        score[[char == letter for char in ans]] = "green"
    return score
```

```python
# Encoding all possible letter-by-letter information that can be returned by the game to the user

def scoring_system(ans):
    return {letter: letter_scoring(ans,letter) for letter in string.ascii_lowercase}
```

```python
# Read in the file that lists all the possible Wordle solutions

f = open("possible_answers.txt","r")
possible_answers = f.read()
possible_answers = possible_answers.split("\n")
f.close()

# Set up a dictionary that has solutions and the corresponding possible
# letter-by-letter information as key-value pairs

poss_ans_encoding = {word:scoring_system(word) for word in possible_answers}
```

```python
# Cut down solution set based on information obtained after a guess

def update_poss_ans_after_info(guess, info, poss_ans):
    for i in range(5):
        poss_ans = [word for word in poss_ans if poss_ans_encoding[word][guess[i]][i] == info[i]]
    return poss_ans
```

```python
# Calculalate the solution size averaged over all possible solutions given the first attempt

def expected_sol_size(guess):
    exp_sol_size, len_poss_ans = 0, len(possible_answers)
    for info in product(["black", "yellow", "green"],repeat=5):
        exp_sol_size += (len(update_poss_ans_after_info(guess, info, possible_answers))**2)/len_poss_ans
    return exp_sol_size
```

```python
# Read in the file that lists all the accepted words in Wordle
# Note that not every 5-letter word is a possible answer in Wordle

f = open("accepted_words.txt","r")
accepted_words = f.read()
accepted_words = accepted_words.split("\n")
f.close()

# Set up a dictionary that has possible guess and the corresponding expectation
# for the subsequent solution size as key-value pairs

# Code is quite time-consuming to run
# Written directly into a file so it can be read on future iterations

filename = "Expected_solution_space_size.txt"
with open(filename, 'w') as file:
    for guess in accepted_words:
        file.write(f"{guess}:{expected_sol_size(guess)}\n")

print(f"Dictionary has been written to {filename}")
```

```python
# Read in the file that lists all the accepted words in Wordle
# Note that not every 5-letter word is a possible answer in Wordle

f = open("accepted_words.txt","r")
accepted_words = f.read()
accepted_words = accepted_words.split("\n")
f.close()

# Set up a dictionary that has possible guess and the corresponding expectation
# for the subsequent solution size as key-value pairs

first_guess_dictionary = {}
for guess in accepted_words:
    print(guess,expected_sol_size(guess))
    first_guess_dictionary.update({guess:expected_sol_size(guess)})

# Code is quite time-consuming to run
# first_guess_dictionary is written into a file so it can be read on future iterations

filename = "Expected_solution_space_size.txt"
with open(filename, 'w') as file:
    for key, value in first_guess_dictionary.items():
        file.write(f"{key}:{value}\n")

print(f"Dictionary has been written to {filename}")
```

```python
# Read in the file with the dictionary that has key-value pairs as the possible guesses
# and corresponding expectation for subsequent solution size

# Filename of the file containing key-value pairs
filename = "Expected_solution_space_size.txt"

# Read in file into an empty dictionary
my_dict = {}
with open(filename, 'r') as file:
    for line in file:
        key, value = line.strip().split(':')
        my_dict[key] = float(value)
```

```python
#Find the word with the smallest expected subsequent solution space
# ie the best word to guess on average
min_value_key, min_value = min(my_dict.items(), key=lambda item: item[1])

print("Key with the minimum value:", min_value_key)
print("Minimum value:", min_value)
```

```python
#Find the 20 words with the smallest expected subsequent solution space 
# and create the corresponding dictionary

print("The smallest 20 values in the dictionary:")
for key, value in sorted(my_dict.items(), key=lambda x: x[1])[:20]:
    print(f"{key}: {value}")
```

```python
# For interest find the word with the largest expected subsequent solution space
# ie the worst word to guess on average
max_value_key, max_value = max(my_dict.items(), key=lambda item: item[1])

print("Key with the maximum value:", max_value_key)
print("Maximum value:", max_value)
```
